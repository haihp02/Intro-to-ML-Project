```python
import numpy as np
import pandas as pd
import preprocess
import pickle
import regex
```

```python
labels = ['Am nhac' 'Am thuc' 'Bat dong san' 'Bong da' 'Chinh tri' 'Chung khoan'
 'Cong nghe' 'Cuoc song do day' 'DROP' 'Doi song' 'Du lich' 'Gia dinh'
 'Giai tri' 'Giao duc' 'Gioi tinh' 'Hinh su' 'Kinh te' 'Lam dep' 'Mua sam'
 'Nghe si' 'Phap luat' 'Phat luat' 'Quan su' 'San khau dien anh'
 'Suc khoe' 'The gioi' 'The gioi tre' 'The thao' 'Thoi su' 'Thoi trang'
 'Van hoa' 'Xa hoi' 'Xe co']
good_labels = ['Bong da', 'Giai tri', 'Du lich', 'Chinh tri', 'Kinh te', 'The thao', 'Nghe si', 'Xe co', 'Suc khoe']

news = pd.read_csv('all.csv')
news
```

```python
# load model
model = pickle.load(open('Data/saved/best_model.sav', 'rb'))

# load feature extractor
feature_extractor = pickle.load(open('Data/saved/feature_extractor.sav', 'rb'))
```

```python
from tqdm import tqdm
```

```python
preds = []
for i in tqdm(range(len(news))):
    ms = news.at[i, 'message'] 
    # Tiền xử lý và loại bỏ stopwords
    data = preprocess.text_preprocessing(ms)
    data = preprocess.remove_stopwords(data)

    # Dự đoán
    np_data = np.array([data])
    feature = feature_extractor.transform(np_data)
    pred = model.predict(feature)
    preds.append(pred)
```

```python
news = pd.read_csv('all.csv')
pred_list = news['pred'].tolist()
```

```python
pred = [pred_list[i][2:-2] for i in range(len(pred_list))]
```
